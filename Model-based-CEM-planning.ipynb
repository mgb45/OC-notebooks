Given running cost $g(x_t,u_t)$ and terminal cost $h(x_T)$ the finite horizon $(t=0 \ldots T)$ optimal control problem seeks to find the optimal control, 
$$u^*_{1:T} = \text{argmin}_{u_{1:T}} L(x_{1:T},u_{1:T})$$ 
$$u^*_{1:T} = \text{argmin}_{u_{1:T}} h(x_T) + \sum_{t=0}^T g(x_t,u_t)$$
subject to the dynamics constraint: $x_{t+1} = f(x_t,u_t)$.

This notebook provides a dirty, brute forcing solution to problems of this form, using the inverted pendulum as an example, and assuming dynamics are not know a-priori. First, we gather state, actions, next state pairs, and use these to train a surrogate neural network dynamics model, $x_{t+1} \sim \hat{f}(x_t,u_t)$, approximating the true dynamics $f$.

We'll then set up a sampling-based optimiser (CEM) to find $$u^*$$ online by rolling out using the surrogate dynamics $\hat{f}$, evaluating the cost, and setting up a Markov chain to converge to this. We'll do this in a continuous control setting, but again no stability guarantees, and using cross-entropy for planning without smoothing the controls will break your robot. Miguel has a great description of CEM: https://jaques.xyz/cem-and-posterior/

```python
class CEMControl:
    
    def __init__(self,dynamics,running_cost,term_cost,u_dim=1,umax=2,horizon=10):
        
        self.dynamics = dynamics
        self.term_cost = term_cost
        self.running_cost = running_cost
        self.horizon = horizon
        self.u_dim = u_dim
        self.umax = umax
        
        self.u_mean = torch.zeros(1,self.horizon)
        self.u_std = torch.ones(1,self.horizon)
        

    def cost(self,x):
        cost = []
        states = [x]
        controls = []
        with torch.no_grad():
            u = self.u_mean + self.u_std*torch.randn(1,self.horizon)
            for j in range(self.horizon-1):
                states.append(self.dynamics(states[-1].reshape(1,-1),self.umax*torch.tanh(u[:,j]).reshape(1,-1)))
                cost.append(self.running_cost(states[-1].reshape(1,-1),self.umax*torch.tanh(u[:,j]).reshape(1,-1)))
                controls.append(u)

        return torch.sum(torch.stack(cost))+self.term_cost(states[-1],u[:,-1]), controls
    
    def minimize(self,xin,Ns=10, Nreps=5):
        
        # Sample a bunch of trials using dynamics and policy
        
        self.u_mean = torch.zeros(1,self.horizon)
        self.u_std = torch.ones(1,self.horizon)
        
        for j in range(Nreps):
            
            losses = []
            controls = []
            for i in range(Ns):
                loss, u = self.cost(xin)

                losses.append(loss)
                controls.append(u[0])

            # Weight actions
            weights = -torch.stack(losses).reshape(-1,)
            weights = weights-torch.min(weights)
            weights = (weights.reshape(-1,1))/torch.sum(weights)
            
            # Recompute mean and std
            dx = torch.stack(controls).squeeze()-self.u_mean
            wdx = weights*dx
            
            self.u_mean = torch.sum(weights*torch.stack(controls).squeeze(),dim=0)
            self.u_std = torch.clamp(torch.sum(wdx*dx,dim=0),1e-4,1)
        
        return self.umax*torch.tanh(self.u_mean)
```
